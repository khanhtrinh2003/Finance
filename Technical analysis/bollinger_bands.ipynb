```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np

#read the data
stock = "GOOGL"
data = yf.download(stock, '2020-01-01', '2023-1-13')

# Calculate the moving average
data['MA'] = data['Close'].rolling(window=20).mean()

# Calculate the standard deviation
data['STD'] = data['Close'].rolling(window=20).std()

# Calculate the upper and lower Bollinger Bands
data['upper_band'] = data['MA'] + 2*data['STD']
data['lower_band'] = data['MA'] - 2*data['STD']

# Plot the Bollinger Bands
plt.plot(data['Close'], label='Close')
plt.plot(data['upper_band'], label='Upper Band')
plt.plot(data['MA'], label='Middle Band')
plt.plot(data['lower_band'], label='Lower Band')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Bollinger Bands')
plt.show()
```
