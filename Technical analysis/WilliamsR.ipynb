```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np

#read the data
stock = "GOOGL"
data = yf.download(stock, '2020-01-01', '2023-1-13')

# Define the period for the Williams %R calculations
period = 14

# Calculate the Williams %R indicator
data['Williams_R'] = -100 * (data['High'].rolling(window=period).max() - data['Close']) / (data['High'].rolling(window=period).max() - data['Low'].rolling(window=period).min())

# Plot the Williams %R indicator
plt.plot(data.index, data['Williams_R'])
plt.xlabel('Time')
plt.ylabel('Williams %R')
plt.title('Williams %R Indicator')
plt.show()
```
