```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance
import numpy as np

#read the data
stock = "GOOGL"
data = yfinance.download(stock, '2020-01-01', '2023-1-13')

# Define the period for the Aroon calculations
period = 25

# Calculate the Aroon Up and Aroon Down
aroon_up = 100*(period - (data['Close'].rolling(window=period).apply(lambda x: np.argmax(x) + 1)))/period
aroon_down = 100*(period - (data['Close'].rolling(window=period).apply(lambda x: np.argmin(x) + 1)))/period

# Plot the Aroon indicator
plt.plot(data.index, aroon_up, label='Aroon Up')
plt.plot(data.index, aroon_down, label='Aroon Down')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Aroon')
plt.title('Aroon Indicator')
plt.show()
```
