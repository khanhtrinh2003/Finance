```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance

#read the data
stock = "GOOGL"
data = yfinance.download(stock, '2020-01-01', '2023-1-13')

# Calculate the ROC indicator
n_periods = 14
data['ROC'] = (data['Close'] - data['Close'].shift(n_periods)) / data['Close'].shift(n_periods)

# Plot the ROC indicator
plt.plot(data.index, data['ROC'])
plt.xlabel('Time')
plt.ylabel('ROC')
plt.title('ROC Indicator')
plt.show()
```
