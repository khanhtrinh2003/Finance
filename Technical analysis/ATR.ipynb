```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

#read the data
stock = "GOOGL"
data = yf.download(stock, '2020-01-01', '2023-1-13')


#calculating the Average True Range
data['ATR'] = data['High'] - data['Low']
data['ATR'] = data['ATR'].shift(1)
data['ATR'].fillna(data['High'] - data['Low'], inplace=True)
data['ATR'] = (data['ATR'] + abs(data['High'] - data['Close'].shift(1)) + abs(data['Low'] - data['Close'].shift(1)))/3
data['ATR'].fillna(data['High'] - data['Low'], inplace=True)
#visualizing the Average True Range
plt.plot(data['ATR'])
plt.title('Average True Range (ATR)')
plt.xlabel('Time')
plt.ylabel('ATR')
plt.show()
```
