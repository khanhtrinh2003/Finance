```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
import yfinance as yf

#read the data
stock = "GOOGL"
data = yf.download(stock, '2022-01-01', '2023-1-13')


# Plot the cand chart
mpf.plot(data, type='candle', style='charles', mav=(3,6,9), volume=True, title='Rising Star Indicator')
```
