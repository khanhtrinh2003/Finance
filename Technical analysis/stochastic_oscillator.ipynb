```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

#read the data
stock = "GOOGL"
df = yf.download(stock, '2020-01-01', '2023-1-13')

# Calculate the stochastic oscillator
# Calculate the rolling mean
rolling_mean = df.Close.rolling(3).mean()

# Calculate the rolling standard deviation
rolling_std = df.Close.rolling(3).std()

# Calculate the %K line
df['%K'] = 100*((df.Close - rolling_mean)/rolling_std)

# Calculate the %D line
df['%D'] = df['%K'].rolling(2).mean()

# Plot the chart
plt.figure(figsize=(12, 8))
plt.plot(df.Close, label='Close Price')
plt.plot(df['%K'], label='%K')
plt.plot(df['%D'], label='%D')
plt.title('Stochastic Oscillator')
plt.legend()
plt.show()
```
