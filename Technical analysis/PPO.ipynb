```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

#read the data
stock = "GOOGL"
df = yf.download(stock, '2020-01-01', '2023-1-13')

# Calculate the 12-day EMA and the 26-day EMA
ema_12 = df.Close.ewm(span=12).mean()
ema_26 = df.Close.ewm(span=26).mean()

# Calculate the Percentage Price Oscillator (PPO)
ppo = 100 * (ema_12 - ema_26) / ema_26

# Plot the data
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(df.index, df.Close, label='Price')
ax.plot(df.index, ema_12, label='EMA 12')
ax.plot(df.index, ema_26, label='EMA 26')
ax.plot(df.index, ppo, label='PPO')
ax.legend(loc='best')
ax.set_title('Percentage Price Oscillator')
ax.set_ylabel('Price in $')
ax.set_xlabel('Date')
plt.show()
```
