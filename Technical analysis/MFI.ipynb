```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

#read the data
stock = "GOOGL"
data = yf.download(stock, '2020-01-01', '2023-1-13')

# Calculate the MFI
typical_price = (data['High'] + data['Low'] + data['Close']) / 3
total_volume = data['Volume']
mfi = typical_price / total_volume

# Visualize the MFI
plt.plot(mfi)
plt.title('Money FLow Index (MFI)')
plt.xlabel('Time')
plt.ylabel('MFI')
plt.show()
```
