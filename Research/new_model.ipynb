```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense

from sklearn.preprocessing import MinMaxScaler
```

```python
tickers="TSLA"
data = pd.DataFrame(yf.download(tickers=tickers, start='2020-01-01', end="2022-12-31")["Close"])
```

```python
# Create 20 days simple moving average column
data['EMA'] = data['Close'].ewm(10).mean()
```

```python
data["diff"]=data["EMA"].diff()
```

```python
d=data[0<data["diff"]][data["diff"]<0.1]
d1=data[-0.1<data["diff"]][data["diff"]<0]
```

```python

plt.plot(data["Close"])
plt.scatter(d.index,d["Close"], color="r")
plt.scatter(d1.index,d1["Close"], color="y")
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
df = scaler.fit_transform(data)
df
```
