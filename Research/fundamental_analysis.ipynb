```python
import yahooquery
import numpy as np
```

```python
ticker = yahooquery.Ticker(['GOOGL', 'AAPL', 'AMZN', "MSFT", "MA"])
```

# Get data

##### Balance sheet

```python
balance_sheet = ticker.balance_sheet()
balance_sheet.columns
```

##### Income statement

```python
income_statement = ticker.income_statement('a')
income_statement.columns
```

##### Cash flow

```python
cash_flow = ticker.cash_flow('a')
cash_flow.columns
```

# Handle data

```python
balance_sheet.dropna(axis=1, inplace=True)
balance_sheet.T
```

```python
AAPL = np.array(balance_sheet.loc["AAPL"].iloc[:,3:].mean())
GOOGL = np.array(balance_sheet.loc["GOOGL"].iloc[:,3:].mean())
MA = np.array(balance_sheet.loc["MA"].iloc[:,3:].mean())
MSFT = np.array(balance_sheet.loc["MSFT"].iloc[:,3:].mean())
AMZN = np.array(balance_sheet.loc["AMZN"].iloc[:,3:].mean())
```

```python
de = np.array([AAPL, GOOGL, AMZN, MSFT])
key = np.array([0.6, 1.7, 5.6, 1.5])
```
