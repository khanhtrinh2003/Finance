```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.linear_model import LinearRegression
```

```python
stock = "GOOGL"
df = pd.DataFrame(yf.download(stock, '2010-01-01', '2022-12-31')["Close"])
```

```python
def rolling(lag, data):
    A = []
    df = data.tolist()
    for i in range(lag, len(df)):
        A.append(df[i-lag:i]) 
    return pd.DataFrame(A)

def LR(x, y):
    model = LinearRegression()
    model.fit(x,y)
    return model

def train_test_split(x,y,test_size):
    i = int(len(x)*test_size)
    j = int(len(y)*test_size)
    return x[:i], x[i:], y[:j], y[j:]
```

```python
lag = 10
A = rolling(lag, df["Close"])
b = df["Close"][lag::]
deon, dete, keyon, keyte = train_test_split(A,b, test_size=0.3)
```

```python
a = LR(deon, keyon)
df1 = pd.DataFrame({"predict": a.predict(dete), "real": keyte})
```

```python
import plotly.graph_objects as go
fig = go.Figure()

# Add first column as a trace
fig.add_trace(go.Scatter(y=df1["predict"], name='predict'))

# Add second column as another trace
fig.add_trace(go.Scatter(y=df1["real"], name='real'))

fig.show()
```
